**Описание датасета**

https://www.kaggle.com/datasets/meirnizri/covid19-dataset



1.   sex: 1 for female and 2 for male.
2.   age: of the patient.
3.   classification: covid test findings. Values 1-3 mean that the patient was diagnosed with covid in different degrees. 4 or higher means that the patient is not a carrier of covid or that the test is inconclusive.
4.   patient type: type of care the patient received in the unit. 1 for returned home and 2 for hospitalization.
5.   pneumonia: whether the patient already have air sacs inflammation or not.
6.   pregnancy: whether the patient is pregnant or not.
7.   diabetes: whether the patient has diabetes or not.
8.   copd: Indicates whether the patient has Chronic obstructive pulmonary disease or not.
9.   asthma: whether the patient has asthma or not.
10.  inmsupr: whether the patient is immunosuppressed or not.
11.  hypertension: whether the patient has hypertension or not.
12.  cardiovascular: whether the patient has heart or blood vessels related disease.
13.  renal chronic: whether the patient has chronic renal disease or not.
14.  other disease: whether the patient has other disease or not.
15.  obesity: whether the patient is obese or not.
15.  tobacco: whether the patient is a tobacco user.
17.  usmr: Indicates whether the patient treated medical units of the first, second or third level.
18.  medical unit: type of institution of the National Health System that provided the care.
19.  intubed: whether the patient was connected to the ventilator.
20.  icu: Indicates whether the patient had been admitted to an Intensive Care Unit.
21.  date died: If the patient died indicate the date of death, and 9999-99-99 otherwise.

Сделаем отдельный параметр, умер ли человек от ковида

```python
def is_died(date):
    if date != '9999-99-99':
        return True
    return False


df1['DEATH'] = df1['DATE_DIED'].apply(is_died)
df1.drop(columns=['DATE_DIED'], inplace=True)
```

Нормализация данных

```python
df1['USMER'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['INTUBED'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['PNEUMONIA'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['PREGNANT'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['DIABETES'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['COPD'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['ASTHMA'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['INMSUPR'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['HIPERTENSION'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['OTHER_DISEASE'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['INMSUPR'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['HIPERTENSION'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['OTHER_DISEASE'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['CARDIOVASCULAR'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['OBESITY'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['RENAL_CHRONIC'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['TOBACCO'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['OBESITY'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1['ICU'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)


df1['PATIENT_TYPE'].replace({1: True, 2: False, 99: None, 97: None, 98: None}, inplace=True)
df1.rename(columns={'PATIENT_TYPE': 'RETURNED_HOME'}, inplace=True)

```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, median_absolute_error, r2_score
```

```python
df3 = df2.dropna(subset = ['AGE','RETURNED_HOME', 'DIABETES', 'COPD', 'ASTHMA','INMSUPR', 'INTUBED','PNEUMONIA' ,'HIPERTENSION','OTHER_DISEASE', 'SEX', 'TOBACCO', 'CLASIFFICATION_FINAL'])
```

```python
Y1 = df3['DEATH']  # Заболевание (DEATH) как целевая переменная
X1=df3[['AGE','RETURNED_HOME', 'DIABETES', 'COPD', 'ASTHMA','INMSUPR', 'INTUBED','PNEUMONIA' ,'HIPERTENSION','OTHER_DISEASE', 'SEX', 'covid_detected', 'TOBACCO', 'CLASIFFICATION_FINAL']] #независимые факторы
X_train, X_test, y_train, y_test = train_test_split(X1, Y1, test_size=0.3, random_state=1)
```

```python
# Логичтическая
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
#масштабирование
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

logreg_model_scaled = LogisticRegression(max_iter=1000)
logreg_model_scaled.fit(X_train_scaled, y_train)#обучение

#Прогнозы
y_pred_scaled = logreg_model_scaled.predict(X_test_scaled)

accuracy_scaled = accuracy_score(y_test, y_pred_scaled)
print('Accuracy=',accuracy_scaled)
print(classification_report(y_test, y_pred_scaled))

```
